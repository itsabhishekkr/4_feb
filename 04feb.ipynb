```python
#Q1)
L=[('Sachin Tendulkar', 34357), ('Ricky Ponting', 27483), ('Jack Kallis', 25534), ('Virat Kohli', 24936)]
def arrang(L):
    R=[]
    l=[]
    for item in L:
        l.append(item[1])
    l.sort()
    for it in L:
        if(it[1]==l[0]):
            R.append(it)
        elif(it[1]==l[1]):
            R.append(it)
        elif(it[1]==l[2]):
            R.append(it)
        elif(it[1]==l[3]):
            R.append(it)
    return(R)
print(arrang(L))
```
